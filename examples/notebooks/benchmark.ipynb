# Neutryx Lab Benchmark
This notebook benchmarks MC vs analytic for BS.

```python
import jax, jax.numpy as jnp
from neutryx.core.engine import MCConfig, simulate_gbm, present_value
from neutryx.models.bs import price as bs_price
key = jax.random.PRNGKey(1)
S,K,T,r,q,sigma = 100.,100.,1.,0.01,0.0,0.2
cfg = MCConfig(steps=252, paths=200_000)
paths = simulate_gbm(key, S, r-q, sigma, T, cfg)
ST = paths[:,-1]
mc = present_value(jnp.maximum(ST-K,0.0), jnp.array(T), r)
an = bs_price(S,K,T,r,q,sigma)
float(mc), float(an)
```
